```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

%matplotlib inline
```

```python
data = pd.read_csv('../baseline/meta.csv')
print(data.head())
```

```python
data['ntime'] = pd.to_datetime(data['time'])
```

```python
data['ntime']
```

```python
from datetime import datetime
```

```python
data['time'] = data['time'].apply(datetime.fromtimestamp)
```

```python
data['time'].dt.hour
```

```python
data['time']
```

```python
import os 
if not os.path.exists('images'):
    os.makedirs('images')
```

```python
plt.figure(figsize=(12,5))
plt.plot(data['time'], data['test_loss'], marker='o')
plt.grid(alpha=0.2)
plt.xlabel('Iteration')
plt.ylabel('Test Loss')
plt.savefig('images/test_loss_over_iters.png', bbox_inches='tight')
```

```python
plt.figure(figsize=(12,5))
plt.plot(data['time'], data['accuracy'], marker='o')
plt.grid(alpha=0.2)
plt.xlabel('Iteration')
plt.ylabel('Accuracy')
plt.savefig('images/accuracy_over_iters.png', bbox_inches='tight')
```

```python
best_loss = [min(data['test_loss'][:i]) for i in range(1,len(data))]
best_acc = [max(data['accuracy'][:i]) for i in range(1,len(data))]
```

```python
plt.figure(figsize=(12,5))
plt.plot(data['time'][1:], best_loss, marker='o')
plt.grid(alpha=0.2)
plt.xlabel('Iteration')
plt.ylabel('Best Loss So Far')
plt.savefig('images/best_loss_so_far_over_iters.png', bbox_inches='tight')
```

```python
plt.figure(figsize=(12,5))
plt.plot(data['time'][1:], best_acc, marker='o')
plt.grid(alpha=0.2)
plt.xlabel('Iteration')
plt.ylabel('Best Accuracy So Far')
plt.savefig('images/best_acc_so_far_over_iters.png', bbox_inches='tight')
```

```python
plt.figure(figsize=(12,12))
plot_vars = ['batch_size', 'learning_rate', 'max_epochs', 'dropout', 'beta1', 'beta2', 'depth', 'init_filters']

for i,plot in enumerate(plot_vars):
    plt.subplot(4,2,i+1)
    plt.plot(data['time'], data[plot], marker='o')
    plt.grid(alpha=0.2)
    plt.ylabel(plot)
    
plt.tight_layout()
plt.savefig('images/parameter_tuning.png', bbox_inches='tight')
```
